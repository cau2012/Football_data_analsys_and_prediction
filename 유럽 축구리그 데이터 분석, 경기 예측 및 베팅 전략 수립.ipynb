---------------------------

### INTRO  

농구, 야구와 달리 축구는 경기 결과를 예측하는 것이 극도로 어렵습니다. 축구는 경기 당 득점이 0~6 점에 불과한 반면 농구와 야구는 경기당 수십 개의 득점이 발생합니다. 때문에 농구와 야구 경기에는 무승부가 거의 발생하지 않습니다. 반면 축구의 경우에는 대략 24.6% 의 경기가 무승부로 종료됩니다. 따라서 경기 결과 예측 문제에서 홈팀을 기준으로 승리, 패배를 이진 분류하는 야구, 농구와는 달리 축구에서는 승리, 무승부, 패배라는 다중 클래스 분류로 접근해야 합니다. 

또한 출루율 을 포함한 몇 가지의 통계 수치만으로 결과 예측을 수행할 수 있는 야구와 달리 축구는 경기장 위의 22명의 선수들이 각자 수행하는 역할과 극도로 체계적인 전술적인 능력, 교체 전술 등에 의해 경기가 결정 됩니다. 이는 수치로 표현하기 어렵습니다. 

스포츠 베팅을 할 경우에는 높은 배당률의 유혹과 독립적인 사건 간 상관관계가 있다고 착각하는 도박사의 오류 등의 인지적 오류에 의해 정확한 경기 결과 예측이 어려워지기도 합니다.

이 프로젝트의 주요 목표는 사람의 주관을 배제하여 축구 경기를 예측하기 위한 모델을 구축하고, 실제 결과에 적용해 보는 것입니다. 프로젝트는 3가지로 구성되어 있습니다. 

첫번째는 탐색적 데이터 분석을 통해 유럽 주요 축구 리그 *(잉글랜드, 스페인, 이탈리아, 독일)* 의 리그별 특성을 확인하고 축구 예측에 사용되는 공식의 유효성을 검증합니다. 또한 28개의 경기력 지표를 T-SNE 알고리즘을 통해 2차원으로 시각화 한 뒤, 유사한 경기력과 전술을 보이는 팀들을 검색해 볼 것입니다.

두번째에서는 여러 머신러닝, 딥러닝 알고리즘, 휴리스틱 방법을 통해 최적의 예측 결과를 도출할 것입니다. 유럽 4개 리그의 2010년부터 2018년까지의 데이터를 사용합니다.

스포츠 베팅에서는 단순히 많이 맞추는 것이 중요한 것이 아니라, 어느 경기에 베팅을 했는지, 얼마나 베팅을 했는지가 더 중요합니다. 세번째에서는 도출해낸 예측 결과를 통해 최적의 수익을 낼 수 있는 스포츠 베팅 전략을 찾아냅니다.

### DATA 

분석 및 예측에 사용된 데이터는 EPL, Laliga, Serie A, Bundesliga 총 4개 리그의 8년 간 11500여개의 경기 데이터입니다. 데이터셋은 다음 웹사이트에서 스크래핑하여 획득하였습니다.

 * https://www.whoscored.com/
 * https://www.transfermarkt.com/
 * http://www.football-data.co.uk/data.php
 
스크래핑에 사용된 라이브러리는 파이썬의 BeautifulSoup 과 Selenium 입니다.

데이터는 크게 경기 외부 수치와 경기 내부 수치로 구분됩니다. 경기 외부 수치는 경기장 위에서 생산된 수치가 아니라 양팀 간의 상대 전적, 연승 혹은 연패 여부, 팀의 몸값, 경기 결과에 대한 스포츠 베팅 업체의 배당률, 부상선수 여부 등 경기가 끝나지 않고도 알 수 있는 수치입니다.

경기 내부 수치는  홈팀과 원정팀의 패스 성공률, 패스 숫자, 패스 위치, 슈팅 숫자, 슈팅 전환율, 태클, 인터셉트 숫자 등 총 28개 입니다.

### FEATURE ENGINEERING  

데이터셋은 경기 외부 수치와 경기 내부 수치(경기력 수치)로 구분되어있지만, 예측하고자하는 경기의 내부 수치는 해당 경기가 종료되기 이전까지는 알 수 없습니다. 때문에 이전 3경기의 경기 수치를 이용해 다음 경기를 예측해야 합니다.

Pandas 라이브러리를 이용하여 feature engineering 을 진행하였습니다. 축구 경기 결과는 홈팀과 원정 팀 중 어느 팀이 일정 수준 이상의 실력을 보이느냐로 결정되는 것이 아니라, 둘 중에 누가 더 잘하는지로 결정 됩니다. 5점차의 승리와 1점차의 승리는 모두 같은 의미를 가지고 있기 때문입니다. 따라서 저는 홈팀과 원정팀의 최근 3경기의 경기력 수치를 상대적인 비율로 표현 하였습니다.
 * H<sub>n</sub> - 홈에서 n 경기 동안의 평균 경기 수치
 * A<sub>n</sub> - 원정에서 n 경기 동안의 평균 경기 수치
 
 H<sub>3</sub> / (H<sub>3</sub> + A<sub>3</sub>)   

이를 통해 경기력을 나타내는 수치는 모두 0에서 1 사이의 값을 가지게 되었으며, 홈 팀과 원정 팀의 절대적인 경기 수치가 아니라 양팀의 상대적인 수치로 표현함으로써 리그별로 구분되는 특성을 완화할 수 있었습니다. 

타겟 데이터는 홈팀 승리(H)는 0, 무승부(D)는 1, 원정팀 승리(A)는 2로 표현됩니다.

* 리그내 예상 경쟁 강도 확인

* 유사한 경기력을 보이는 팀 찾기

8개 시즌, 4개 리그에 참여하는 팀의 숫자는 총 624개입니다. 매 시즌에 지속적으로 참여하는 팀이라고 할지라도 시즌마다 감독이 달라지기도 하고 새로 영입하거나 방출한 선수에 따라 전술이 바뀌기 때문에 경기력의 수치가 달라집니다. 때문에 매 시즌마다 꾸준히 리그에 참여하는 같은 팀이라고 할지라도 시즌이 달라지면 다른 팀으로 구분하였습니다. 이를 통해 특정 시즌의 특정 팀과 가장 유사한 팀들을 찾아내는 함수를 만들었습니다.

우선, Pandas의 groupby 함수를 이용하여 시즌 별 각 팀의 28개 경기력 수치의 평균을 구한 뒤, T-SNE 알고리즘을 이용하여 이를 2차원의 공간으로 바꾸었습니다. 2차원 공간에서 특정 팀을 선택하면, 유클리드 거리를 기반으로 가장 가까운 N개의 팀을 출력합니다.

* ARTIFICIAL NEURAL NETWORK

머신러닝 모델에 적용한 것처럼 ANN 모델에도 Oversampling 과정을 거친 데이터를 적용했습니다. 분데스리가에 대한 예측력은 매우 낮은것을 확인하여 모델의 학습, 평가에 분데스리가 데이터를 모두 제외했습니다. ANN의 파라미터는 다음과 같습니다. 

>* Learning rate : 0.0001
* Batch size : 200
* Epochs : 150
* Drop out : 0.1

ANN 모델에 데이터를 학습시킬때 염두에 둔 것은, 단순히 최대한 높은 accuracy 를 획득하는 것이 아니라, 3개 클래스 중 가장 예측하기 어려운 무승부를 얼마나 많이 예측하는지였습니다. 모델이 무승부에 대한 특성을 잘 학습하면서도 일정 수준 이상의 accuracy를 가질 수 있도록 Learning rate 를 최대한 낮추었습니다. ANN 모델은 Pytorch 라이브러리로 제작되었고, 128개 노드를 가진 13개의 Affine 레이어로 구성되어 있습니다.

### BUILDING A BETTING STRATEGY

베팅 전문가들은 스포츠 베팅을 수익 수단이 아니라 놀이 수단으로 여기라고 조언 합니다. 스포츠 베팅으로 지속적인 수익을 얻는 것은 매우 힘든일이고, 스포츠 경기에 베팅을 하는 것 만으로도 스포츠 관람에 대한 즐거움이 증대되기 때문입니다. 하지만 베팅에서 돈을 잃는 것은 유쾌한 경험이 아니기에 많은 사람들은 베팅에서의 수익을 극대화하기 위해 나름의 방법을 고안합니다. 자신이 고안한 방법으로 수익을 얻는 사람들도 있지만 지속적으로 베팅에 참여한 사람들 대부분은 잃은 돈이 얻은 돈보다 많습니다. 

베팅의 횟수나 베팅 금액, 베팅 성향에 따라 달라지겠지만, 베팅 회사를 상대로 돈을 잃지 않기 위해서는 52.4% 의 경기를 맞춰야한다고 합니다. (https://www.thesportsgeek.com/sports-betting/math/) 그러나 지속적으로 베팅을 하면서 돈을 잃지 않고, 더 나아가 수익을 얻기 위해서는 많이 맞추는 것만이 중요한 것이 아닙니다. 수익률(배당률)과 지금까지 축적 된 자신의 적중률(적중 확률)을 바탕으로 최선의 수익을 낼 수 있는 몇개의 경기에 베팅을 해야합니다. 

예를들어 스페인 La Liga 에서 25회의 우승 경험을 가지고 있는 바르셀로나의 2010년 9월 11일 홈 경기의 경우, 바르셀로나의 승리에 대한 배당률이 1.06에 불과하고 바르셀로나 패배에 대한 배당률은 34나 될 정도로 오즈메이커들은 바르셀로나의 승리를 압도적으로 예측하였습니다. 하지만 결과는 바르셀로나의 패배였고, 바르셀로나가 패할 것이라고 예측한 사람들은 베팅액의 3300%(3400% - 100%)나 되는 수익을 얻은데 반해 바르셀로나에 베팅을 한 사람들은 6%의 수익을 얻으려다가 베팅액을 잃고 말았습니다. 6%의 수익은 바르셀로나가 질수도 있다는 위험부담을 감수하기엔 너무 적은 수익이었습니다.

켈리 공식을 이용한다면 이런 실수를 막을 수 있습니다. 포트폴리오 분산 투자를 위한 용도로도 사용 되는 켈리 공식을 스포츠 베팅에 적용한다면, 경기 예측에 성공할 확률과 기대 수익을 고려하여 어느 경기에 얼마나 베팅할 것인지 결정할 수 있습니다. 켈리 공식은 다음과 같습니다.

>(BP - Q)/B  

> B = 소수점 배당률-1   
 P = 성공 확률  
 Q = 실패 확률 (1-P)  

이 공식을 상기 제시한 바르셀로나 경기*(배당률: 홈승리:1.06 무승부:11 홈패배:34)*에 적용하면 다음과 같은 값이 나옵니다. 성공 확률은 ANN 모델의 Recall 값*(홈 승리-0.62, 무승부-0.32, 홈 패배-0.57)*으로 적용합니다.

> * 바르셀로나가 승리할 것이라고 예측하는 경우  
> {(1.06-1) * 0.62 - (1-0.62)}/(1.06-1) = -5.7133  

> * 바르셀로나가 패배할 것이라고 예측하는 경우  
> {(34-1) * 0.57 - (1-0.57)}/(34-1) = 0.5569  

> * 바르셀로나가 비길 것이라고 예측하는 경우  
> {(11-1) * 0.32 - (1-0.32)}/(11-1) = 0.252  

켈리 값은 가지고 있는 자산에서 어느 정도 만큼 해당 경기에 베팅하라고 지시해줍니다. 만약 켈리 값이 0 이하이면, 그 경기에는 베팅을 하지 않아야 합니다. 켈리 값은 해당 경기에 대해 바르셀로나의 패배에 가지고 있는 자산의 55.69%를 베팅하라고 말 합니다.

### RESULTS

이번 프로젝트에선 2010년부터 2018년까지의 4개 유럽 축구 리그 데이터를 분석하고 경기 결과를 예측하였으며 그 결과를 바탕으로 직접적인 수익 창출을 위한 시뮬레이션을 진행하였습니다. 

데이터 분석 과정에서 얻게 된 여러 통찰 중 하나는, 라리가 리그의 경우에는 실제 리그 내 격차에 비해 오즈메이커들이 계산하는 리그 내 격차가 다른 리그에 비해 상대적으로 과장되어있다는 것입니다. 비록 라리가 승점의 표준편차가 다른 리그보다 상대적으로 높은 시즌이 몇 보이지만, 오즈메이커들이 예상하는 리그내 격차 만큼 압도적이지는 않습니다. 이는 대체로 오즈메이커들이 라리가 내 강팀의 승리를 강하게 예상하는데 반해, 실제로는 그 예상만큼 라리가 내 강팀이 압도적으로 강하지 않기 때문에 생기는 현상이라고 판단하고 있습니다. 때문에 주식에서 수익을 내는 주요 방법 중 하나인 저평가 주식 발굴처럼 실력에 비해 가치가 낮게 평가 받고 있는 라리가 중하위권 팀에 대해 주의 깊게 찾아보는 것이 베팅 수익 상승에 도움을 줄 것이라고 예상됩니다. 

또한 T-SNE 를 통해 데이터의 feature를 44차원에서 2차원으로 변환한 뒤, 유클리드 거리로 측정한 가장 유사한(가장 가까운 거리)몇개의 과거 경기 결과만으로도 미래의 경기를 예측하는 것이 가능하다는 것을 확인 하였습니다. 0.4568 의 정확도는 아주 뛰어난 성능이라고 할 수는 없지만, 홈팀의 승,무,패 군집에 예측하고자하는 경기가 해당되어있는 것을 확인하는 것만으로도 미래의 경기에 대한 예측에 도움을 줄 수 있을것 입니다.

머신러닝, 딥러닝으로 경기를 예측하는 과정에서는 각 리그마다 정확도가 다르다는 것을 확인했습니다. Serie A의 경우 경기의 58~60%를 맞춘데 반해, Bundesliga 는 불과 48~49%의 경기만을 적중했습니다. 이는 데이터 표현 문제일수도 있고, 학습을 시킨 모델이 로버스트하지 못한 것이 원인일 수도 있을 것입니다. 또한 Oversampling을 통해 홈 팀 승리 예측에 대한 편향을 줄이고 축구 경기 결과에서 가장 예측하기 어려운 무승부 예측의 Recall 값을 높이고자 하였으나 만족할만한 성과를 내지 못했습니다. parameter 를 아무리 조정하더라고 하더라도, 무승부 결과의 데이터와 원정팀 승리(홈팀 패배)의 데이터의 형태가 비슷하기 때문에 모델이 이를 잘 구분해내지 못하였습니다. 

축구 경기에는 질 경기를 비기게 만들고, 비길 경기를 이기게 만들어주는 특별한 선수들이 있습니다. 그런 선수들의 능력을 데이터에 반영하지 않은 것도 모델의 성능 향상에 한계점으로 작용했습니다. 하지만 제한적인 데이터로 도출해낸 머신러닝, 딥러닝의 55% 정확도는 웬만한 축구 전문가가 보이는 정확도로써, 축구 경기 베팅에 있어서 충분히 수익을 낼 수 있음을 확인하였습니다.
